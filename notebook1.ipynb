# Procedimento para instalação da linguagem Python  
-----
<h2>Entendendo o que é o anaconda </h2> 
<p>Anaconda é uma distribuição completamente gratis do Python.Inclui mais de 400 pacotes,os mais populares pacotes do python: math,science,engineering e data analysis.</p> 
<h2>Instalação:</h2>
<p><h4>Linux:</h4> 
<li>Download do instalador (dependendo da arquitetura do seu computador instale 32 bits ou 64 bits): https://www.continuum.io/downloads </li> 
<li>Abra o terminal</li>   
Python 2.7
<br> Digite esse código
<div style="background-color:black; color:white;"> >> bash Anaconda2-2.5.0-Linux-x86_64.sh </div> <br>  
Python 3.5
<br> Digite esse código
<div style="background-color:black; color:white;"> >> bash Anaconda3-2.5.0-Linux-x86_64.sh </div> <br>

<font style="font-weight:bold;">obs</font>: Use o comendo "Bash" mesmo se você não estiver usando o bash sheel <br>
<font style="font-weight:bold;">obs</font>: Verifique qual versão você baixou para proceder com o nome correto do arquivo 
</p> 

<p><h4>OS X:</h4>
<li>Download do instalador (dependendo da arquitetura do seu computador instale 32 bits ou 64 bits): https://www.continuum.io/downloads </li>
<li>Click duplo no arquivo .pkg para instalar e siga as instruções</li>

<p><h4>Windows:</h4>
<li>Download do instalador (dependendo da arquitetura do seu computador instale 32 bits ou 64 bits): https://www.continuum.io/downloads </li>
<li>Click duplo no arquivo.exe para instalar e siga as instruções</li>
</p> 

<h2>Configurando o ambiente python</h2> 
<p>Antes de começarmos a usar, abra(execute) o anaconda, ao entrar no ambiente anaconda insira esse comando :</p>
<div style="background-color:black; color:white;"> >> conda update conda </div>

# Procedimento para instalação da linguagem Julia 
----

<h2>Entendendo um pouco mais sobre Julia</h2> 
<p>Julia é uma linguagem de programação de alto nível e de alta performace para computação tecnica, possuindo uma sintex muito familiar para o usuário.</p> 

<h2>Instalação</h2> 
<p><h4>Linux</h4> <br> 
<li>Julia disponibiliza um PPA (Personal Package Archive) para os sistemas Linux que permite fazer updating automaticamente da ultima versão estavel da linguaem Julia</li> 
<li>Para usar PPA basta fazer os seguintes comandos:</li> <br>
<div style="background-color:black; color:white;"> >> sudo add-apt-repository ppa:staticfloat/juliareleases <br>
>> sudo add-apt-repository ppa:staticfloat/julia-deps <br> 
>> sudo apt-get update <br> 
>> sudo apt-get install julia
</div>
</p> 

<p> 
<h4>OS X</h4> <br> 
Para o Mac, o Julia-version.dmg (download:http://julialang.org/downloads/ ) ele é que contem o Julia.app. A instalação é do mesmo jeito que se instal qualquer software no Mac - copia o Julia.app para o seu hard-driver (pode colocar em qualquer canto) ou rodaatraves da imagem do disco.
</p> 

<p> 
<h4>Windows</h4> <br> 
<li>Baixe o Julia.exepara a sua plataforma: http://julialang.org/downloads/</li>
</p> 

<h2>Instalando o IJulia</h2> 
<p> 
Para poder instalar o IJulia basta seguir os seguintes passos <br> 
<li>Abra o terminal executanado o Julia ou abra o Julia.exe</li> <br> 
Digite o seguinte código <br>
<div style="background-color:black; color:white;"> >> Pkg.add("IJulia") <br> 
>> using IJulia <br> 
>> notebook() <br></div>
</p>

# Instalando o Git e Github 
----  
<h2>Instalando o git</h2>
<h4>Git para o Linux</h4> 
<p>Para instalar o Git no Linux basta: <br> 
**1**-Abra o terminal <br> 
**2**-Digite a seguinte linha de comando: 
<div style="background-color:black; color:white;"> >> sudo apt-get install git</div>
</p> 
<h4>Git para OS X</h4> 
<p>
Basta instalar usando a forma gráfica, download : http://sourceforge.net/projects/git-osx-installer/
</p>
<h4>Git para Windows</h4> 
<p>
Basta fazer o download na página: http://msysgit.github.com
</p>  

<h2>Usando o github</h2> 
<p>Para podere ter acesso ao github (repositório online e publico), basta fazer a conta no site - https://github.com/ - <br>
Após obter uma conta no github, basta criar um repositório e seguir os passos que o site fornece
</p>
